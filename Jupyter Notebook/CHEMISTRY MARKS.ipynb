```python
from camelot import read_pdf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fhandle=read_pdf("resources/pdf/SomeBasicConceptsOfChemistry(Responses)2.pdf",pages='all')
```

```python
page=[]
for n in fhandle:
    y=n.df
    y.drop([0],inplace=True)
    page.append(y)
    
d=pd.concat(page,ignore_index=True)
```

##  COPY

```python
df1=d.copy()
```

```python
df1.columns=["Score(10)","Name","Class","Combination"]
```

```python
df1.head()
```

## COPY

```python
df2=df1.copy()
```

```python
df1[df1["Name"]==""]
```

```python
df1.iloc[[34,66,68,122,126,186]]
```

```python
df2.Name[df2.Name=='']=df2['Score(10)']
```

```python
df2=df1.replace({"Class":""},np.NaN)
df2.dropna(subset=["Class"],inplace=True)
```

```python
df2["Score(10)"]=df2["Name"].replace(r'[A-Za-z. ]',"",regex=True)
df2.Name=df2.Name.replace(r'[0-9/ ]','',regex=True)
df2["Score(10)"]=df2["Score(10)"].replace(r'/10',"",regex=True)
```

```python
df2.head(10)
```

```python
df2.iloc[[34,66,68,122,126,186]]
```

```python
df2.Name=df2.Name.str.lower()
```

```python
df2.loc[df2['Score(10)']=='','Score(10)']=0
```

```python
df2.loc[df2['Score(10)']==np.NaN,'Score(10)']=0
```

```python
df2.iloc[[34,66,68,122,126]]
```

```python
df2['Score(10)']=df2['Score(10)'].apply(pd.to_numeric)
```

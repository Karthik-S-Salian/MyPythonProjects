We shall examine the effect of all factors individually

```python
# function to plot
df_left=df[df.left==1]
df_notleft=df[df.left==0]
def plot_relation(column):
    dl=df_left[column].value_counts(sort=False)
    dnl=df_notleft[column].value_counts(sort=False)
    plt.scatter(dl.index.values,dl.values,label='left',color='r')
    plt.scatter(dnl.index.values,dnl.values,color='b',label='not left')
    plt.xlabel(column)
    plt.ylabel('no of employee')
    plt.title(f'Effect of {column} on employees leave decision')
    plt.legend(loc='upper right')
    plt.show()
```

1. No of projects

```python
plot_relation('number_project')
```

Employees who made less then 3 projects are most who left

2. Satisfactory Level

```python
plot_relation('satisfaction_level')
```

we can say that employee with satisfaction level less that 0.44 are more likely to leave which is  same as give by mean

## DATA PREPROCESSING

creating dummy columns for salary

```python
processed_df = pd.get_dummies(df,columns=['salary'], prefix="salary")
processed_df.drop(['salary_high','left','Work_accident','Department'],axis='columns',inplace=True)
processed_df.head(2)
```

Standardizing data

```python
sc = StandardScaler()
X = sc.fit_transform(processed_df)
```

```python
#target variable
y=df.left.values
```

## Model Building

```python
models = [LogisticRegression(max_iter=1000),DecisionTreeClassifier(criterion='entropy')
          ,RandomForestClassifier(max_features='log2'),SVC(C=10)]

score=[]
for model in models:
    score.append(cross_val_score(model,X,y).mean())
```

```python
model_name=['Logistic Regression','Decision Tree','Random Forest','SVM']
model_stats=pd.DataFrame(index=model_name)
model_stats['Average score']=score
model_stats
```

```python
model_params={
    'svm':{
        'model':SVC(C=10),
        'params':{
            'kernel':['rbf','linear'],
            'gamma':['scale', 'auto']
        } 
    },
    'DecisionTreeClassifier':{
        'model':DecisionTreeClassifier(),
        'params':{
            'criterion':['gini','entropy']
            'max_features':['auto',"log2"]
        }   
    },
    'RandomForestClassifier':{
        'model':RandomForestClassifier(),
        'params':{
            'criterion':['gini','entropy'],
            'max_features':['auto',"log2"]
        } 
    } 
}
```

```python
from sklearn.model_selection import GridSearchCV

score=[]
for name,model in model_params.items():
    clf=GridSearchCV(model['model'],model['params'],cv=5,return_train_score=False)
    clf.fit(X,y)
    score.append({
        'model':name,
        'best_score':clf.best_score_,
        'best_params': clf.best_params_
    })
pd.DataFrame(score)
```
